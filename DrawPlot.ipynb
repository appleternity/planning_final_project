```python
# Python Libraries
import json
import os
import collections

# 3rd party lib
import numpy as np
import matplotlib.pyplot as plt

# list all files in directory
log_dir = '/Users/liushi-hao/Workspace/planning_final_project/log-probability'
fs = os.listdir(log_dir) 
prefix_type = collections.defaultdict(list)

for filename in fs:
    prefix, tp = filename.split('-')
    prefix_type[prefix].append(tp)
    
# draw for each group 
for prefix, tps in prefix_type.items():
    for tp in tps:
        x, y = [], []
        abs_filename = os.path.join(log_dir, prefix + '-' + tp)
        with open(abs_filename, 'r') as f:
            lines = f.readlines()

        # load json
        try:
            for line in lines:
                data = json.loads(line)
                x.append(data['e'])
                y.append(data['goal'])
        except ValueError:  # includes simplejson.decoder.JSONDecodeError
            print('Decoding JSON has failed')

        # draw line
        plt.plot(x, y, label=tp)

    # display graph
    plt.xlabel('iterations')
    plt.ylabel('goal')
    plt.title(prefix)
    plt.legend()
    plt.show()
```
